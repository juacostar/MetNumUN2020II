Una función de Taylor para una función $f(x)$ en uan dimensión sobre un punto $x=a$ está dada por:

$f(x) \approx f(a) + f^{'}(a)(x-a) + \frac{1}{2!}f^{''}a(x-a)^2 +...$

* Es decir:
$\sum_{n=0}^\infty = \frac{f^n(a)}{n!}(x-a)^n$


### Calculando lña serie de Taylor para $f(x)=cos(x), f(x)=sen(x)$ y $f(x)=e^x$
* Siguiendo la definición para $f(x)=cos(x)$ se tiene que:
\
$cos(x) \approx cos(a) - sen(a)(x-a) - \frac{1}{2}cos(a)(x-a)^2 + \frac{1}{6}sen(a)(x-a)^3 $
* Para $f(x)=sen(x)$:\
$sen(x) \approx sen(a) + cos(a)(x-a) + \frac{1}{2}sen(a)(x-a)^2 - \frac{1}{6}cos(a)(x-a)^3$
* Para $f(x)=e^x$ :\
$e^x \approx e^a(x-a) + \frac{1}{2}e^a(x-a)^2 + \frac{1}{6}cos(a)(x-a)^3$

Cuando asumimos $x=a=0$ se tiene que:
* Para $f(x)=cos(x)$:\
$cos(x)=1 - \frac{1}{2}x^2$
* Para $f(x)=sen(x)$:\
$sen(x)=x - \frac{1}{6}x^3$
* Para $f(x)=e^x$:\
$e^x=1 + x + \frac{1}{2}x^2 + \frac{1}{6}x^3$

Ejecutando código Python para obtener el valor de dichas series se tiene que:

```
from sympy import series, Symbol
from sympy.functions import sin, cos, exp,log
from sympy.plotting import plot
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = 13,10
plt.rcParams['lines.linewidth'] = 2
```

```
# Define symbol
x = Symbol('x')
```

```
# Function for Taylor Series Expansion

def taylor(function, x0, n):
    """
    Parameter "function" is our function which we want to approximate
    "x0" is the point where to approximate
    "n" is the order of approximation
    """
    return function.series(x,x0,n).removeO()
```

```
print('sin(x) =', taylor(sin(x), 0, 4))

print('cos(x) =', taylor(cos(x), 0, 4))

print('e(x) =', taylor(exp(x), 0, 4))
```

Evaluando dichas aproximaciones se tiene que:

```
print('sin(1) =', taylor(sin(x), 0, 4).subs(x,1))

print('cos(1) =', taylor(cos(x), 0, 4).subs(x,1))

print('e(1) =', taylor(exp(x), 0, 4).subs(x,1))
```

### Expansión de la serie de Taylor de $f(x)=sen(x)$

```
# This will plot sine and its Taylor approximations

p = plot(sin(x),taylor(sin(x),0,1),taylor(sin(x),0,3),taylor(sin(x),0,5),
         (x,-3.5,3.5),legend=True, show=False, xlim=(-5,5), ylim=(-2,2))

p[0].line_color = 'blue'
p[1].line_color = 'green'
p[2].line_color = 'firebrick'
p[3].line_color = 'black'
p.title = 'Taylor Series Expansion for Sine'
p.show()
```

### Expansión de la serie de Taylor de $f(x)=cos(x)$

```
# This will plot cosine and its Taylor approximations
p = plot(cos(x),taylor(cos(x),0,2),taylor(cos(x),0,4),taylor(cos(x),0,6),
         (x,-4.5,4.5),legend=True, show=False, xlim = (-5,5), ylim=(-2,2))

p[0].line_color = 'blue'
p[1].line_color = 'green'
p[2].line_color = 'firebrick'
p[3].line_color = 'black'
p.title = 'Taylor Series Expansion for Cosine'
p.show()
```

### Expansión de la serie de Taylor de $f(x)=e^x$

```
# This will plot exponent and its Taylor approximations

p = plot(exp(x),taylor(exp(x),0,1),taylor(exp(x),0,2),taylor(exp(x),0,3),
         (x,-2,2),legend=True, show=False, xlim=(-5,5), ylim=(-2,2))

p[0].line_color = 'blue'
p[1].line_color = 'green'
p[2].line_color = 'firebrick'
p[3].line_color = 'black'
p.title = 'Taylor Series Expansion for Exponent'
p.show()
```

Siguiendo la definición de la serie de Taylor para otras funciones se tiene que:
* Para $f(x) = \frac{1}{1-x}$: \
$f(x) = \frac{1}{1-x} = \sum_{k=0}^{\infty} x^k = 1 + x + x^2 + x^3 + ...$
* Para $f(x)=ln(1+x)$\
$f(x)=ln(1+x) = \sum_{k=1}^\infty(-1)^{k-1}\frac{x^k}{k} = x - \frac{x^2}{2} + \frac{x^3}{3} - ...$
* Para $f(x)=ln(x)$\
$f(x)=ln(x) = \sum_{k=1}^\infty(-1)^{k-1}\frac{(x-1)^k}{k} = x - \frac{x^2}{2} + \frac{x^3}{3} - ...$

Usando las expansiones de cada función se tiene que:

#### Para $f(x) = \frac{1}{1-x}$:

```
p = plot(1/(1-x),taylor(1/(1-x),0,1),taylor(1/(1-x),0,2),taylor(1/(1-x),0,3),
         (x,-2,2),legend=True, show=False, xlim=(-5,5), ylim=(-2,2))

p[0].line_color = 'blue'
p[1].line_color = 'green'
p[2].line_color = 'firebrick'
p[3].line_color = 'black'
p.title = 'Taylor Series Expansion for 1/(1-x)'
p.show()
```
