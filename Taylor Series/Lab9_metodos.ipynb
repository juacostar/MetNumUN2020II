```
import matplotlib.pyplot as plt
import numpy as np
```

```
extent = 0.0, 4.0, 0.0, 4.0 
   
width = 400
height= 400
pixels = np.full((width,height,3),0)
for i in range(width):
  for j in range(height):
    if (j>i+100): 
      pixels[height-j-1,i,0] = pixels[height-j-1,i,1] = pixels[height-j-1,i,2] = 255
      #print(i,j)

plt.title('bisection_convergence')
plt.imshow(pixels,extent = extent)
plt.show()
```

```
def func(x): 
    return (x-1) *(x-2) *(x-3)*(x-4)


x = np.linspace(0.0,5.0,500)
fig, ax = plt.subplots()
ax.plot(x,func(x))
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.grid(True, which='both')
plt.title('f(x)=(x-1) *(x-2) *(x-3)*(x-4)')
plt.xlabel('x')
plt.ylabel('f(x)')
```

Para realizar la convergencia de raíces del polinomio mediante el método de bisección se concluyeron los diferentes intervalos para los cuáles convergen:
* Para el intervalo $[-\infty, x]|1 < x < 2 $ converge para la raíz $x=1$, dado a que en este intervalo se presenta cambio de signo el cual es necesario para hacer efectivo este método, y en este intervalo se encuentra la raíz mencionada.
* Para el intervalo $(1,x]|2 < x < 3$, converge a la segund raíz $x=2$. Debido a que también se presenta cambio de signo y se encuentra la raíz mencionada.
*  Para el intervalo $(2,x]|3 < x < 4$, converge a la tercera raíz $x=3$. Se presenta cambio de signo y se encuentra dicha raíz.
* Para el intervalo $(3,x]|x > 4$ Converge a la cuarta raíz $x=4$. En este intervalo se presenta la raíz y se presenta tambipen cambio de signo.
Sin embargo, existen intervalos en los cuales no se puede realizar la convergencia a ninguna raíz, como el intervalo $[ 2 , 3 ]$,en dondeno existe cambio de signo y por tanto no se puede relizar el método de bisección. Tmabién aplica para los intervalos $[-\infty, 1]$ y $[4,\infty]$.

```
width = height = 80
xi=0.0
xf=5.0
eps = 1e-3
mxiter = 1e4

pixels = np.full((width,height,3),0)
# roots
r1 = 1
r2 = 2
r3 = 3
r4 = 4
   
multcol = 5

for i in range(height):
            
            aa =   float(xi + (i * (xf-xi) /height))
            
            for j in range(width):
                bb= float(xi + (j* (xf-xi)/ width))

                a = aa
                b = bb
                c=a
                iter = 0
                
                while ((b-a) >= eps and (func(a) * func(b) <= 0)): 
                   
                    # Check if a is root 
                    if (func(a) == 0.0): 
                      c=a
                      break

                    # Check if b is root 
                    if (func(b) == 0.0): 
                      c=b
                      break
  
                    # Find middle point 
                    c = (a+b)/2
    
                    # Decide the side to repeat the steps 
                    if (func(c)*func(a) < 0): 
                        b = c 
                    else: 
                        a = c 
                    
                    iter = iter + 1
                    
                    if (iter > mxiter):
                     break

                # Pixels colored using the roots
                if abs(c-r1)<eps:
                    # color red
                    pixels[height-j-1,i] = (255 - iter*multcol, 0, 0)
                elif abs(c-r2)<=eps:
                    # color green
                    pixels[height-j-1,i] = (0, 255 - iter*multcol, 0)
                elif abs(c-r3)<=eps:
                    # color blue
                    pixels[height-j-1,i] = (0, 0, 255 - iter*multcol)
                elif abs(c-r4)<=eps:
                    # color black
                    pixels[height-j-1,i] = (0,  iter*multcol,0)
                elif (b < a):
                    pixels[height-j-1,i] = (255, 255, 0)
                elif (func(a) * func(b) >= 0):
                    pixels[height-j-1,i] = (255, 255, 255)
                else:   
                    pixels[height-j-1,i] = (0, 0, 0)

                
     
plt.title('bisection_convergence')
plt.xlabel("a in ("+str(xi)+","+str(xf)+")")
plt.ylabel("b in ("+str(xi)+","+str(xf)+")")
extent = xi, xf, xi, xf
plt.imshow(pixels,extent =extent)
plt.show()
```
